```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
dataset = pd.read_csv('sat.csv')
```

```python
type(dataset)
```

```python
dataset
```

```python
# linearity check
```

```python
x1 = dataset['SAT']
```

```python
x1
```

```python
y = dataset ['GPA']
```

```python
y
```

```python
plt.scatter (x1,y)
plt.xlabel ('SAT')
plt.ylabel ('GPA')
plt.show ()
```

```python
import statsmodels.api as sm
```

```python
x = sm.add_constant(x1)
```

```python
x
```

```python
model = sm.OLS(y,x).fit()
```

```python
model.summary()
```

```python
test = pd.DataFrame(data=[1600,1950,1500],columns=['SAT'])
```

```python
test
```

```python
test = sm.add_constant(test)
```

```python
test
```

```python
model.predict(test)
```
